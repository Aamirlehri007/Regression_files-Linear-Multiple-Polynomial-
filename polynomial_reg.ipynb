```python
# Another important example

import numpy 
import pandas as pd
import matplotlib.pyplot as plt

# Importing the dataset
dataset= pd.read_csv("Position_Salaries.csv")

x1 = dataset.iloc[:, 1:2].values
y1 = dataset.iloc[:, 2].values
```

```python
dataset.head()
```

```python
# splitting the dataset into the Training and Test set
from sklearn.model_selection import train_test_split 
x_train, x_test, y_train, y_test = train_test_split(x1,y1, test_size=0.2, random_state=0)
```

```python
# Fitting Linear Regression to the dataset
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(x1,y1)

# Visualization the Linear regression result
def viz_linear():
    plt.scatter(x1,y1,color= "red")
    plt.plot(x1, lin_reg.predict(x1),color="blue")
    plt.title("Truth or Bluff (Linear Regression)")
    plt.xlabel("Position Level")
    plt.ylabel("Salary")
    plt.show()
    return
viz_linear()  
```

```python
# Fitting polynomial Regression  to the dataset
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree=4)
x1_poly = poly_reg.fit_transform(x1)
pol_reg= LinearRegression()
pol_reg.fit(x1_poly, y1)

# Visualization the polynomial Regression result
def viz_polynomial():
    plt.scatter(x1,y1,color= "red")
    plt.plot(x1, pol_reg.predict(poly_reg.fit_transform(x1)),color="blue")
    plt.title("Truth or Bluff (Polynomial Regression)")
    plt.xlabel("Position Level")
    plt.ylabel("Salary")
    plt.show()
    return
viz_polynomial()
```

```python
# predicting a new result with Linear Regression 
pred_linear= lin_reg.predict([[11]])
```

```python
pred_polynomial = pol_reg.predict(poly_reg.fit_transform([[11]]))
```

```python
print("Linear Regression results   =", pred_linear)
print("Polynomial Regression results   =", pred_polynomial)

print("The difference is           =", pred_linear-pred_polynomial)

```
